```python
video=cv2.VideoCapture(0)
Model_face=cv2.CascadeClassifier("trained model/haarcascade_frontalface_alt.xml")
Model_eye=cv2.CascadeClassifier("trained model\haarcascade_eye.xml")
while True:
    flag,img=video.read()  # if img exist then flag value will be True else flag value will be false
    if flag==False:
        break

    img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face=Model_face.detectMultiScale(img2)
    cv2.putText(img,text="Press 'c' to stop",org=(10,20),fontFace=cv2.FONT_HERSHEY_COMPLEX,color=(0,0,255),thickness=2,fontScale=1)
    for x,y,w,h in face:
        cv2.rectangle(img,(x,y,w,h),(0,255,0),2)
        

        img3=img2[y:y+h,x:x+w]
        eye=Model_eye.detectMultiScale(img3)
        for ex,ey,ew,eh in eye:
            cv2.rectangle(img,(x+ex,y+ey,ew,eh),(0,255,0),2)

    cv2.imshow("wind",img)
    key=cv2.waitKey(10)
    
    if key==ord("c"):
        break
cv2.destroyAllWindows()
video.release()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_sc=sc.fit_transform(X)
print(X_sc)
from sklearn.decomposition import PCA
pca=PCA(0.99)
X_pca=pca.fit_transform(X_sc)

from sklearn.ensemble import AdaBoostClassifier
Model=AdaBoostClassifier()   # untrained model
Model.fit(X_pca,y)
Model.score(X_pca,y)
```

```python
import cv2
vdo=cv2.VideoCapture(0)
Model_face=cv2.CascadeClassifier(r"trained model\haarcascade_frontalface_alt_tree.xml")
i=1
while True:
    flag,img=vdo.read()
    if flag==False:
        break
    img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face=Model_face.detectMultiScale(img2)
    for x,y,w,h in face:
        img2=cv2.resize(img,(100,100))
        img2=img2.flatten()
        img2=sc.transform([img2])
        img2=pca.transform(img2)
        result=Model.predict(img2)
        cv2.putText(img,text=result[0],org=(10,10),fontFace=cv2.FONT_HERSHEY_COMPLEX,color=(0,0,255),thickness=2,fontScale=1)
        cv2.rectangle(img,(x,y,w,h),(0,255,0),2)
        
    cv2.imshow("wind",img)
    key=cv2.waitKey(20)
    if key==ord("c"):
        break
cv2.destroyAllWindows()
vdo.release()
```

```python
import cv2
import numpy as np

video=cv2.VideoCapture(0)
Model_face=cv2.CascadeClassifier("trained model/haarcascade_frontalface_alt.xml")
while True:
    flag,img=video.read()  # if img exist then flag value will be True else flag value will be false
    if flag==False:
        break

    img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)   #black & white img
    face=Model_face.detectMultiScale(img2)
    # cv2.putText(img,text="Press 'c' to stop",org=(10,20),fontFace=cv2.FONT_HERSHEY_COMPLEX,color=(0,0,255),thickness=2,fontScale=1)
    for x,y,w,h in face:
        cv2.rectangle(img,(x,y,w,h),(0,255,0),2)
        img3=img2[y:y+h,x:x+w]   #slice the img
        img_resize=cv2.resize(img3,dsize=(100,100))   #resize the img
        img_flatten=img_resize.flatten() 
        img_sc=sc.transform([img_flatten])
        img_pca=pca.transform(img_sc)
        result=Model.predict(img_pca)
        cv2.putText(img,text=result[0],org=(x+10,y+20),fontFace=cv2.FONT_HERSHEY_COMPLEX,color=(0,0,255),thickness=2,fontScale=1)
        result2=Model.predict_proba(img_pca)
        cv2.putText(img,text=f"{np.max(result2)}",org=(x+10,y+40),fontFace=cv2.FONT_HERSHEY_COMPLEX,color=(0,0,255),thickness=2,fontScale=1)
    cv2.imshow("wind",img)
    key=cv2.waitKey(10)
    
    if key==ord("c"):
        break
cv2.destroyAllWindows()
video.release()
```
