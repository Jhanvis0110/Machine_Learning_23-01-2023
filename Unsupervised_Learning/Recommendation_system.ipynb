# Similarity Score

Cosine similarity lies between -1 to 1

```python
# Step1:  creating data on small scale first
s1="I learn java"
s2="I learn python"
s3="Virth Kohalu is cricketer"
corpus=[s1,s2,s3]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Step2: Filtering data means converting data into corpus into small letters and removing stop words
tf=TfidfVectorizer(lowercase=True,stop_words='english')
```

```python
# Step3:  sparse matrix to  dense matrix conversion
sparse_matrix=tf.fit_transform(corpus)
dense_matrix=sparse_matrix.toarray()
```

```python
# Step4: Cosine similarity help us to get the distance between vectors using math's cos angle, where cos 90 means 0 mean not match b/w 2 vectors and cos0 means
# which mean it match 100% to each other.
# In cosine similarity, cos degree value lies between -1 to +1 where -1 is absolutely oppostive and +1 means absolutly same.
# Note: cosine similarity show similarity when degree is bigger.

from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(dense_matrix)
cs
```

```python
# Step5: Cosine distance help to find the distance b/w 2 vectors or more. If distance is higher that means not match than
       # lower that means both vectors are same.
# Note: cosine distance show similarity when distance is smaller.

from sklearn.metrics.pairwise import cosine_distances
cd=cosine_distances(dense_matrix)
cd
```

```python
# Step9:  conversion of description of movies of dataframe into corpus using pandas df1.description.
# Step10: TfidVectorizer and countvectorizer both use for creating feature from corpus nd also contain corpus preprocessing.
# Step11:  Creating dense matrix from sparse matrix with the help of toarray() method.
```

```python
import re
corpus4=[]
for i in corpus3:
    corpus4.append(re.sub("[^a-z]","",i.lower()))
corpus4
```

```python
from nltk.corpus import stopwords

sw=stopwords.words("english")
sw.remove("not")
```

```python
len(sw)
```

```python
for i in sw:
    if "not" == i:
        print(True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(stop_words=sw)
dense_matrix=tf.fit_transform(corpus4).toarray()
```

```python
# Step12:  importing algo for recommendation which is nearestneighbours it partically work like knn(knn doesnot have cosine: codine_similarity and cosine_distance)

# Most import NearestNeighbours is for "unsupervised learning" where target is not required

from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors()             #bydefault it calculate equilidian distance.
model.fit(dense_matrix)
```

```python
model.kneighbors([dense_matrix[0]],n_neighbors=4)     # yha predict ni hota neighbours puchte h
```

```python
name=[i.lower() for i in list(df2.name)]
file=input("Enter film name:")
index_film=name.index(file)

similarity_score,film_list=model.kneighbors([dense_matrix[index_film]],n_neighbors=5)

for i in film_list[0][1:]:
    print(name[i])
```
