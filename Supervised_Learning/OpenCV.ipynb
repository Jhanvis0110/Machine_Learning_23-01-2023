```python
import cv2
Model_face=cv2.CascadeClassifier("trained model\haarcascade_frontalface_alt.xml")
img=cv2.imread("sachin-tendulkar.webp")
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
face=Model_face.detectMultiScale(img2)
for x,y,w,h in face:           # coordinates of img
    cv2.rectangle(img,(40,40),(150,170),(0,255,0),2)
cv2.imshow("wind",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
Model_face=cv2.CascadeClassifier("trained model\haarcascade_frontalface_alt.xml")
img=cv2.imread("group1.jpg")
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
face=Model_face.detectMultiScale(img2,minNeighbors=14)   
for x,y,w,h in face:
    cv2.rectangle(img,(x,y,w,h),(0,255,0),2)
cv2.imshow("wind",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
Model_face=cv2.CascadeClassifier("trained model\haarcascade_frontalface_alt.xml")
Model_eye=cv2.CascadeClassifier("trained model\haarcascade_eye.xml")
img=cv2.imread("group1.jpg")
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
face=Model_face.detectMultiScale(img2)
eye=Model_eye.detectMultiScale(img2)
for x,y,w,h in face:
    cv2.rectangle(img,(x,y,w,h),(0,255,0),2)

for ex,ey,ew,eh in eye:
    cv2.rectangle(img,(ex,ey,ew,eh),(0,255,0),2)
cv2.imshow("wind",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
Model_face=cv2.CascadeClassifier("trained model\haarcascade_frontalface_alt.xml")
Model_eye=cv2.CascadeClassifier("trained model\haarcascade_eye.xml")
img=cv2.imread("group1.jpg")
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
face=Model_face.detectMultiScale(img2,minNeighbors=14)

for x,y,w,h in face:
    cv2.rectangle(img,(x,y,w,h),(0,255,0),2)

    img3=img2[y:y+h,x:x+w]
    eye=Model_eye.detectMultiScale(img3)
    for ex,ey,ew,eh in eye:
       cv2.rectangle(img,(x+ex,y+ey,ew,eh),(0,255,0),2)

cv2.imshow("wind",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2

video=cv2.VideoCapture("Naina Ke Teer ( Rani Ho Tera Laya Mein Lal Sharara ) Renuka Panwar Tanu Rawat Vivek Swati Vikram.mp4")

while True:
    flag,img=video.read()  # if img exist then flag value will be True else flag value will be false
    if flag==False:
        break
    Model_face=cv2.CascadeClassifier("trained model/haarcascade_frontalface_alt.xml")
    Model_eye=cv2.CascadeClassifier("trained model\haarcascade_eye.xml")
    img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face=Model_face.detectMultiScale(img2,minNeighbors=14)
    cv2.putText(img,text="Press 'c' to stop",org=(10,20),fontFace=cv2.FONT_HERSHEY_COMPLEX,color=(0,0,255),thickness=2,fontScale=1)
    for x,y,w,h in face:
        cv2.rectangle(img,(x,y,w,h),(0,255,0),2)
        

        img3=img2[y:y+h,x:x+w]
        eye=Model_eye.detectMultiScale(img3)
        for ex,ey,ew,eh in eye:
            cv2.rectangle(img,(x+ex,y+ey,ew,eh),(0,255,0),2)

    cv2.imshow("wind",img)
    key=cv2.waitKey(10)
    
    if key==ord("c"):
        break
cv2.destroyAllWindows()
video.release()
```
