```python
review1="food is good and food is tasty"     #document
review2="foo8889d is!$ not tasty"
review3="cheap quality&&&&"
review4="expensive** stuff#"
review5="Food is aweso0me^^"


corpus=[review1,review2,review3,review4,review5] # corpus is not feature
target=[1,0,0,0,1]
corpus
```

```python
def punctuation_remover(corpus):
    corpus1=[]
    for i in corpus:
         jhanvi=""
         for j in i:
             if j not in "1234567890!#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~":
                jhanvi+=j

         corpus1.append(jhanvi.lower()) 
    return corpus1
```

```python
import string
string.punctuation
```

```python
from nltk.corpus import stopwords
len(stopwords.words("english"))
sw=stopwords.words("english")
sw.remove("not")
sw.remove("no")
```

```python
def stopword_remover(corpus1):
    corpus2=[]
    for i in corpus1:
        split_word=i.split()
        new_str=""
        for j in split_word:
            if j not in sw:
                new_str=new_str+j+" "
        result=corpus2.append(new_str.strip())
    return corpus2
```

```python
print(corpus)
def cleaner(corpus):
     result=punctuation_remover(corpus)
     clean_corpus=stopword_remover(result)
     return clean_corpus
corpus1=cleaner(corpus)
corpus1

```

```python
# creating feature using countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
cv.fit_transform(corpus1)
```

```python
print(cv.fit_transform(corpus1))
print(cv.fit_transform(corpus1).toarray())      #dense matrix
print("all feature name :",cv.get_feature_names_out())
```

```python
X=cv.fit_transform(corpus1).toarray()  #feature
X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf=TfidfVectorizer(ngram_range=(1,2))     # ngram increases the feature

sparse_metric=tf.fit_transform(coprus_review_cleaned)
X=sparse_metric.toarray()
len(tf.get_feature_names_out())
from sklearn.naive_bayes import BernoulliNB
Model2=BernoulliNB()
Model2.fit(X,y)

sample2=["foo3252545d quali35435ty is 66546not so good"]
sample3=cleaner(sample2)
print(sample3)
sample4=tf.transform(sample3).toarray()
print(sample4)
print(Model2.predict(sample4))
Model2.predict_proba(sample4)
```
